# Reformat data into a single `.dat` file

### open data

```python
import pandas as pd

# gives the notebook access to the cluster_expansion personal library
from sys import path
path.append('/Users/prabh/200 RESEARCH/Cluster Expansion Project')
import cluster_expansion as ce
```

```python
chrV = pd.read_csv('ChrVData/ChrVLibrary.txt', sep = '\t')
chrV.columns = chrV.columns.str.strip()
chrV = chrV.drop(columns=['Sequence #'])

chrV_free = pd.read_csv('ChrVData/ChrVReadcounts.txt', sep = '\t')
chrV_free.columns = chrV_free.columns.str.strip()

tiling = pd.read_csv('TilingData/tilingLibrary.txt', sep = '\t')
tiling.columns = tiling.columns.str.strip()
tiling = tiling.drop(columns=['Sequence #'])

tiling_free = pd.read_csv('TilingData/tilingReadcounts.txt', sep = '\t')
tiling_free.columns = tiling_free.columns.str.strip()
```

```python
display_on = True

def display_head(df, num_rows, on):
  if (on):
    display(df.head(num_rows))

display_head(df=chrV, num_rows=2, on=display_on)
display_head(df=chrV_free, num_rows=2, on=display_on)
display_head(df=tiling, num_rows=2, on=display_on)
display_head(df=tiling_free, num_rows=2, on=display_on)
```

##### Rename Columns

```python
chrV.rename(columns={
  'Cyclizability (n=26)':'C26',
  'Cyclizability (n=29)':'C29',
  'Cyclizability (n=31)':'C31'
}, inplace=True)

chrV_free.rename(columns={
  'C26 c_n':'C26 count',
  'C26 c_s':'C26 count sample',
  'C29 c_n':'C29 count',
  'C29 c_s':'C29 count sample',
  'C31 c_n':'C31 count',
  'C31 c_s':'C31 count sample',
}, inplace=True)

tiling.rename(columns={
  'Cyclizability (n=26)':'C26',
  'Cyclizability (n=29)':'C29',
  'Cyclizability (n=31)':'C31'
}, inplace=True)

tiling_free.rename(columns={
  'C26 c_n':'C26 count',
  'C26 c_s':'C26 count sample',
  'C29 c_n':'C29 count',
  'C29 c_s':'C29 count sample',
  'C31 c_n':'C31 count',
  'C31 c_s':'C31 count sample',
}, inplace=True)

if display_head:
  display(chrV.head(2))
  display(chrV_free.head(2))
  print('\n')
  display(tiling.head(2))
  display(tiling_free.head(2))
```

### Trim sequences

```python
# Before
def check_length(df):
  try:
    bool_mask = df['Sequence'].apply(lambda s: isinstance(s, str) and len(s) == 50)
    if (~bool_mask).any():
      display(df.loc[~bool_mask].head())
    return bool_mask.all()
  except Exception as e:
    print("Malformed dataframe", e)

print(f"chrV Sequences length 50: {check_length(chrV)}")
print(f"tiling Sequences length 50: {check_length(tiling)}")
```

```python
chrV['Sequence'] = chrV['Sequence'].apply(ce.trim)
tiling['Sequence'] = tiling['Sequence'].apply(ce.trim)

# After
print(f"chrV Sequences length 50: {check_length(chrV)}")
print(f"tiling Sequences length 50: {check_length(tiling)}")
```

### Remove digesion motifs

##### Count total digestion motifs

```python
def countMotifs(df):
  try:
    numMotifs = 0
    numMotifs += df['Sequence'].str.contains('GAAGAGC').sum()
    numMotifs += df['Sequence'].str.contains('GCTCTTC').sum()
    return numMotifs
  except Exception as e:
    print('Malformed dataframe', e)
    return 0
```

```python
# Count digesion motifs
print(f'# chrV digestion motifs: {countMotifs(chrV)}')

# checking for digestion motifs in the tiling Data
print(f'# tiling digestion motifs: {countMotifs(tiling)}')
```

##### Remove digestion motifs

```python
# remove all digestion motifs from the ChrVData
chrV = chrV[~chrV['Sequence'].str.contains('GAAGAGC|GCTCTTC')].reset_index(drop=True)
tiling = tiling[~tiling['Sequence'].str.contains('GAAGAGC|GCTCTTC')].reset_index(drop=True)

# Count digesion motifs
print(f'# chrV digestion motifs: {countMotifs(chrV)}')

# checking for digestion motifs in the tiling Data
print(f'# tiling digestion motifs: {countMotifs(tiling)}')
```

##### Check that sequences arrays are now equal

```python
chrV_isequal = chrV['Sequence'] == chrV_free['Sequence']
tiling_isequal = tiling['Sequence'] == tiling_free['Sequence']

print(f'chrV sequences are equal: {all(chrV_isequal)}')
print(f'tiling sequences are equal: {all(tiling_isequal)}')
```

### Drop duplicate sequences

```python
def count_duplicates(df):
  try:
    original_count = len(df)
    unique_count = df['Sequence'].nunique()
    duplicates = original_count - unique_count
    return duplicates
  except Exception as e:
    print("Malformed dataframe", e)
    return 0

print(f'# duplicates in chrV: {count_duplicates(chrV)}')
print(f'# duplicates in chrV_free: {count_duplicates(chrV_free)}')

print(f'# duplicates in tiling: {count_duplicates(tiling)}')
print(f'# duplicates in tilint_free: {count_duplicates(tiling_free)}')
```

```python
chrV = chrV.drop_duplicates(subset='Sequence', keep='first').reset_index(drop=True)
chrV_free = chrV_free.drop_duplicates(subset='Sequence', keep='first').reset_index(drop=True)

tiling = tiling.drop_duplicates(subset='Sequence', keep='first').reset_index(drop=True)
tiling_free = tiling_free.drop_duplicates(subset='Sequence', keep='first').reset_index(drop=True)

print(f'# duplicates in chrV: {count_duplicates(chrV)}')
print(f'# duplicates in chrV_free: {count_duplicates(chrV_free)}')

print(f'# duplicates in tiling: {count_duplicates(tiling)}')
print(f'# duplicates in tilint_free: {count_duplicates(tiling_free)}')

chrV_isequal = chrV['Sequence'] == chrV_free['Sequence']
tiling_isequal = tiling['Sequence'] == tiling_free['Sequence']

print(f'chrV sequences are equal: {all(chrV_isequal)}')
print(f'tiling sequences are equal: {all(tiling_isequal)}')
```

```python
display_head(df=chrV, num_rows=2, on=display_on)
display_head(df=chrV_free, num_rows=2, on=display_on)
display_head(df=tiling, num_rows=2, on=display_on)
display_head(df=tiling_free, num_rows=2, on=display_on)
```

### Merge dataframes

```python
chrV_merged = chrV.merge(chrV_free, on = 'Sequence', how = 'left', suffixes = ('', '_rc')).reset_index(drop=True)
tiling_merged = tiling.merge(tiling_free, on = 'Sequence', how = 'left', suffixes = ('', '_rc')).reset_index(drop=True)
display_head(chrV_merged, num_rows=2, on=display_on)
display_head(tiling_merged, num_rows=2, on=display_on)
```

##### Reorder dataframe

```python
ordering = [
  'Sequence',
  'C0free',
  'C0',
  'C26', 'C26 count', 'C26 count sample',
  'C29', 'C29 count', 'C29 count sample',
  'C31', 'C31 count', 'C31 count sample',
  'Amplitude', 'Phase'
]

try:
  chrV_merged = chrV_merged[ordering]
  tiling_merged = tiling_merged[ordering]
except Exception as e:
  print('Failed to reorder columns', e)
```

```python
display_head(df=chrV_merged, num_rows=2, on=display_on)
print(len(chrV_merged))
display_head(df=tiling_merged, num_rows=2, on=display_on)
print(len(tiling_merged))
```

### Save to a `.dat` file

```python
save = False
if save:
  chrV_merged.to_csv('ChrV.dat', sep = '\t', index=False)
  tiling_merged.to_csv('Tiling.dat', sep = '\t', index=False)
```
